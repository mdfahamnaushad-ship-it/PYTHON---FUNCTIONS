## Theoretical Questions

##Q1 What is the difference between a function and a method in Python?
- A function in Python is a block of code that performs a specific task and can be called anywhere in the program. A method, on the other hand, is similar but is always associated with an object or class (like append() for lists). Simply put, all methods are functions, but not all functions are methods.

##Q2 Explain the concept of function arguments and parameters in Python.
- Parameters are the variables defined in a function’s definition, while arguments are the actual values passed when calling that function. For example, in def greet(name):, name is a parameter, and when you call greet("Faham"), “Faham” is the argument.

##Q3 What are the different ways to define and call a function in Python?
- Functions can be defined using the def keyword or by using a lambda expression for short, unnamed functions. You can call a function simply by writing its name followed by parentheses and passing arguments if required, like add(5, 3).

##Q4 What is the purpose of the return statement in a Python function?
- The return statement sends a value back from the function to the caller. It allows the function to output results that can be used later in the program. Without return, a function only performs an action but does not give back any value.

##Q5 What are iterators in Python and how do they differ from iterables?
- An iterator is an object that allows sequential access to elements, one at a time, using next(). An iterable is any object that can produce an iterator, like lists or strings. In simple terms, lists are iterables, and iter(list) creates an iterator from them.

##Q6 Explain the concept of generators in Python and how they are defined.
- Generators are special functions that yield one item at a time using the yield keyword instead of return. They help save memory because they don’t store all values at once; instead, they generate them on the fly when needed.

##Q7 What are the advantages of using generators over regular functions?
- Generators are memory efficient because they produce data only when required. They are also useful when dealing with large datasets or streams of data since they don’t load everything into memory at once. This makes programs faster and more efficient.

##Q8 What is a lambda function in Python and when is it typically used?
- A lambda function is a short, anonymous function written in one line using the lambda keyword. It’s often used for simple operations that don’t need a full function definition, like sorting or filtering small datasets.

##Q9 Explain the purpose and usage of the map() function in Python.
- The map() function applies a given function to all items in an iterable (like a list) and returns a new iterator with the results. It’s commonly used for performing the same operation on every element, such as converting all numbers in a list to their squares.

##Q10 What is the difference between map(), reduce(), and filter() functions in Python?
- map() applies a function to each item, filter() selects items based on a condition, and reduce() combines items into a single value (like summing them). In short, map() transforms, filter() selects, and reduce() combines.

##Q1 Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def sum_even(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

print(sum_even([10, 15, 20, 25, 30]))
```

##Q2 Create a Python function that accepts a string and returns the reverse of that string.

```python
def reverse_string(s):
    return s[::-1]

print(reverse_string("Faham"))
```

##Q3 Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

```python
def squares(numbers):
    return [n**2 for n in numbers]

print(squares([2, 4, 6, 8]))
```

##Q4 Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

for n in range(1, 201):
    if is_prime(n):
        print(n, end=" ")
```

##Q5 Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        self.count += 1
        value = self.a
        self.a, self.b = self.b, self.a + self.b
        return value

for num in Fibonacci(10):
    print(num, end=" ")
```

##Q6 Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power_of_two(limit):
    for i in range(limit + 1):
        yield 2 ** i

for val in power_of_two(5):
    print(val)
```

##Q7 Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()
```

##Q8 Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

```python
data = [(1, 4), (3, 2), (5, 1), (2, 3)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```

##Q9 Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit.

```python
celsius = [0, 20, 30, 40]
fahrenheit = list(map(lambda c: (9/5)*c + 32, celsius))
print(fahrenheit)
```

##Q10. Create a Python program that uses filter() to remove all the vowels from a given string.

```python
text = "Hello Faham from New Delhi"
result = ''.join(filter(lambda ch: ch.lower() not in 'aeiou', text))
print(result)
```
